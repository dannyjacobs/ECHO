Purpose:

Determine whether the drone  or its payload contributes to any noise detected by the spectrum analyzer during flight testing at the Sevilleta LWA. Test equipment used was a Fieldfox spectrum analyzer, biconical antenna, and low noise amplifier. Background noise spectra was recorded before each flight during the field test, and several spectra were taken during each flight as the drone completed its flight path. Each flight spectra was compared to its respective background to determine if there is an excess of noise above the background, and if any of the noise is being emitted into the band of interest, 70MHz.

Characterizing the Background Noise: 

Each day the drone completed several flight paths. Prior to the start of each flight path a spectra of the background was taken with the drone, payload, and any other heavily emitting electronics (aside from the LWA antennas) turned off. This spectra would be used as a basis to compare to the spectra taken during the flight path when the drone and payload are turned on. A recurring problem was the noise floor of the background occasionally being higher than the noise floor during the flight which made it difficult for comparison. On the third day only three of the 12 flight spectra taken match the noise floor of the associated background. Day four had better results with 11 of the 17 flights having matching noise floors.

To begin, all of the main flight background spectra were collected into an array and a for loop was used to iterate through each spectra file and plotting them.

Final Confidence Bound Code

A function was created to iterate through each spectra in the array, skip the first several lines, and append the values in columns 0 and 1 (frequency and power columns) into another array which is returned.

```python
def load_fieldfox(filename):
    f=open(filename)
    lines=f.readlines()
    d=[]
    for line in lines:
        if line[0]=='!': continue
        if line.startswith('BEGIN') or line.startswith('END') :continue
        vals=line.split(',')    
        d.append([vals[0],vals[1]])
    return np.array(d,dtype=np.float)

```

```python
bgrounds_files=[r'C:\Users\shani\Documents\Work\LoCo Labs\ECHO\Sevilleta\Day 3 10-25\DAY3BACK2.csv',
         r'C:\Users\shani\Documents\Work\LoCo Labs\ECHO\Sevilleta\Day 3 10-25\F6_B.csv',
        r'C:\Users\shani\Documents\Work\LoCo Labs\ECHO\Sevilleta\Day 3 10-25\F7_B.csv',
         r'C:\Users\shani\Documents\Work\LoCo Labs\ECHO\Sevilleta\Day 4 10-26\D4FB.csv',
        r'C:\Users\shani\Documents\Work\LoCo Labs\ECHO\Sevilleta\Day 4 10-26\D4FB_2.csv' ]
```

We use a for loop that utilizes the function to go through each file except the first and append the values to the the array bgrounds. 'bgrounds' is then stacked along axis -1 and sorted along axis -2

```python
bgrounds_data=[]
for bgfile in bgrounds_files[1:]: #[1:] skips the first file, which is a spectra with the highest noise floor
    
    bgrounds_data.append(load_fieldfox(bgfile)) #

bgrounds_data=np.stack(bgrounds_data, axis=-1)

bgsorted1=np.sort(bgrounds_data, axis=2)
```

'bgrounds' is then plotted in a figure one by one in multi colors and the confidence bound is plotted in black. When analyzing each background together, it is found that 80% of them have power emissions at ~-100 dBm or lower, as is shown in the following figure:

```python
fig=figure()
fig = figure(figsize=(20,10))
plot(bgrounds_data[:,0,:]/1e6, bgrounds_data[:,1,:]) #all bground plots plot(x=freq,y=pwr)
#plot(bgrounds[:,1,:]) #[dim1,dim2,dim3], 1-power in 2nd dim. also plots bgrounds but x=#ofpnts
#plot(bgsorted[:,1,-2], 'k', label='80% Confidence Bound') #black line=80% confidence interval. x=#points
plot(bgsorted1[:,0,-2]/1e6,bgsorted1[:,1,-2], 'k', label='Confidence Bound')

fig.suptitle('Day 3 and 4 Backgrounds', fontsize=20)
xlabel('Frequency in MHz', fontsize=15)
ylabel('Power in dBm', fontsize=15)
legend(fontsize=10)
grid()
```
